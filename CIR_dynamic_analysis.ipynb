```python
j = 2009
for key in df_list.keys():
    colName = []
    colName.append(str(j))
    if j == 2009:
        df_new = df_list[key]
        df_new = df_new.iloc[2:18,12].to_frame()
        df_new.columns = colName[:]
    else:
        df_new[colName] = df_list[key].iloc[2:18,12].to_frame()
    j = j + 1
```

```python
df_new.loc['Чистий дохід']=df_new.loc['процентні доходи']+df_new.loc['комісійні доходи']+df_new.loc['результат від торговельних операцій']+df_new.loc['інші операційні доходи']-df_new.loc['процентні витрати']-df_new.loc['комісійні витрати']
df_new.loc['Адміністративні та інші операційні витрати']=df_new.loc['інші операційні витрати']+df_new.loc['загальні адміністративні витрати']
df_new.loc['CIR']=df_new.loc['Адміністративні та інші операційні витрати']/df_new.loc['Чистий дохід']
df_new.loc['Зміна CIR'] = df_new.loc['CIR'].diff().fillna(0)
```

```python
import seaborn as sns
from matplotlib import colors
import numpy as np
cm = sns.diverging_palette(5, 250, as_cmap=True)

def background_gradient(s, m, M, cmap='PuBu', low=0, high=0):
    rng = M - m
    norm = colors.Normalize(m - (rng * low),
                            M + (rng * high))
    normed = norm(s.values)
    c = [colors.rgb2hex(x) for x in plt.cm.get_cmap(cmap)(normed)]
    return ['background-color: %s' % color for color in c]

even_range = np.max([np.abs(impact_coef.iloc[:,1:].values.min()), np.abs(impact_coef.iloc[:,1:].values.max())])

impact_coef.iloc[:,1:].style.apply(background_gradient,
               cmap=cm,
               m=-even_range,
               M=even_range).set_precision(2)

# impact_coef.iloc[1:,:].style.background_gradient(cmap='coolwarm')
```
